```python
def plot_decisiion_boundry(estimator, X,y):
    estimator.fit(X,y)
    X_color = X.sample(21000)
    y_color = y.loc[X_color.index]
    y_color = y_color.map(lambda r:'red' if r==1 else 'yellow')
    
    x_axis,y_axis = np.arange(0,1, 0.005), np.arange(0,1,0.005)
    xx,yy = np.meshgrid(x_axis,y_axis)
    xx_ravel = xx.ravel()
    yy_ravel = yy.ravel()
    
    x_grid = pd.DataFrame([xx_ravel,yy_ravel]).T

    y_grid_prediction = SVC().fit(X,y).predict(x_grid)
    y_grid_prediction = y_grid_prediction.reshape(xx.shape)
    
    fig,ax = plt.subplots(figsize=(10,10))
    ax.contourf(xx,yy, y_grid_prediction, cmap=plt.cm.autumn_r, alpha=0.5)
    ax.contourf(X_color.iloc[:,0], X_color.iloc[:,1], color=y_color, alpha=1)
    
```
